```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# SÃ©lectionner uniquement la colonne "MEDV" et les autres colonnes
corr_df = df.corr()[["MEDV"]]

# Afficher la matrice de corrÃ©lation avec une heatmap
sns.set(style="ticks")
plt.style.use("dark_background")

plt.figure(figsize=(12, 8))
sns.heatmap(corr_df, annot=True, cmap='coolwarm')
plt.title("CorrÃ©lation avec la colonne MEDV")
plt.show()
```

```python

# Tracer le scatter plot pour la variable LSTAT
scatter = plt.scatter(df.LSTAT, df.MEDV, c=df.RM, cmap='cool')
# Ajouter des labels et un titre
plt.colorbar(scatter, label='Nombre moyen de chambres (RM)')
plt.xlabel('% Population de statut infÃ©rieur')
plt.ylabel('Valeur mÃ©diane des maisons')
plt.title('Relation entre le % de population de statut infÃ©rieur et la valeur mÃ©diane des maisons')
# Afficher le graphique
plt.show()

```
